```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
theta_1 = np.linspace(-np.pi,np.pi,20)
theta_2 = np.linspace(-np.pi,np.pi,20)
l1 = 1
l2 = 1

randomsamples = 20

data = []
for theta_1_i in theta_1:
    for theta_2_i in theta_2:
        x = np.round(l1*np.cos(theta_1_i) + l2*np.cos(theta_2_i),2)
        y = np.round(l1*np.sin(theta_1_i) + l2*np.sin(theta_2_i),2)
        inputposition = (x,y)
        outputjoints = (np.round(theta_1_i,2),np.round(theta_2_i,2))
        for sample in range(randomsamples):
            inputjoints = (np.round(theta_1_i + 0.2*(np.random.random()-0.5) ,2) ,np.round(theta_2_i + 0.2*(np.random.random()-0.5) ,2))
            data.append( [ inputposition , inputjoints , outputjoints ] )
```

```python
s = np.array(data)

fig = plt.figure()
plt.xlim([-np.pi,np.pi])
plt.ylim([-np.pi,np.pi])
plt.scatter( s[:,1,0], s[:,1,1], color = 'red')
plt.scatter( s[:,2,0], s[:,2,1], color = 'blue')
plt.show()
```

```python
fig = plt.figure()
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.scatter( s[:,0,0], s[:,0,1], color = 'blue')
plt.show()
```

```python
# open file for writing, "w" is writing
mydata = csv.writer(open("dataset.csv", "w"))

mydata.writerow([ 'input position', 'input joint', 'output joint' ])

# loop over data
for data_i in data:

    # write data elements to file
    mydata.writerow([ data_i[0], data_i[1], data_i[2] ])
```
